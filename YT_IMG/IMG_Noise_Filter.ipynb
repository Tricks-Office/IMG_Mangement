### 이미지 읽고 노이즈 필터 적용하기

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from tkinter import filedialog

f_selected = filedialog.askopenfile(title = 'Image Test')
f_selected = f_selected.name.replace('/','\\')

# 이미지 로드
input_image_path = f_selected
output_image_path_gaussian = 'denoised_image_gaussian.jpg'
output_image_path_median = 'denoised_image_median.jpg'
output_image_path_bilateral = 'denoised_image_bilateral.jpg'

img = cv2.imread(input_image_path, cv2.IMREAD_COLOR)

# 원본 이미지 출력
plt.figure(figsize=(10, 10))
plt.subplot(221), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')

# Gaussian 블러링
gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)
cv2.imwrite(output_image_path_gaussian, gaussian_blur)
plt.subplot(222), plt.imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)), plt.title('Gaussian Blur')

# Median 블러링
median_blur = cv2.medianBlur(img, 5)
cv2.imwrite(output_image_path_median, median_blur)
plt.subplot(223), plt.imshow(cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)), plt.title('Median Blur')

# Bilateral 필터링
bilateral_filter = cv2.bilateralFilter(img, 9, 75, 75)
cv2.imwrite(output_image_path_bilateral, bilateral_filter)
plt.subplot(224), plt.imshow(cv2.cvtColor(bilateral_filter, cv2.COLOR_BGR2RGB)), plt.title('Bilateral Filter')

plt.show()
```
